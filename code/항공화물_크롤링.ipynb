```python
# 필요 패키지
import selenium
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from tqdm import tqdm_notebook
import time
```

```python
date_list = []
for y in range(2018, 2023):
    start_month = 1
    end_month = 13
    if y == 2022:
        end_month = 4
    for m in range(start_month, end_month):
        if len(str(m)) == 1:
            m = '0' + str(m)
        date_list.append(str(y) + '-' + str(m))
date_list
```

```python
area_country = [['미주','미국'], ['유럽','독일'], ['유럽','프랑스'], ['유럽','이탈리아'], ['유럽','러시아'], ['중국','중국'], ['아시아','베트남'], ['아시아','싱가포르'], ['아시아','말레이시아'], ['일본', '일본'], ['아시아', '홍콩']]
```

```python
url = 'https://www.airportal.go.kr/knowledge/statsnew/air/area.jsp#'
```

```python
driver = webdriver.Chrome('./driver/chromedriver')
driver.get(url)
```

```python
res = []
for s in area_country:
    for date in date_list:
        # 시작 연도 선택
        xpath = '//*[@id="start_year"]'
        start_year = Select(driver.find_element(By.XPATH,xpath))
        start_year.select_by_visible_text(str(int(date[0:4])))
        # 시작 월 선택
        xpath = '//*[@id="start_month"]'
        start_month = Select(driver.find_element(By.XPATH,xpath))
        start_month.select_by_visible_text(str(int(date[5:7])))

        # 끝 연도 선택
        xpath = '//*[@id="end_year"]'
        end_year = Select(driver.find_element(By.XPATH,xpath))
        end_year.select_by_visible_text(str(int(date[0:4])))
        # 끝 월 선택
        xpath = '//*[@id="end_month"]'
        end_month = Select(driver.find_element(By.XPATH,xpath))
        end_month.select_by_visible_text(str(int(date[5:7])))

        # 지역 선택
        xpath = '//*[@id="areaGubun"]'
        area = Select(driver.find_element(By.XPATH,xpath))
        area.select_by_visible_text(s[0])

        # 국가 선택
        xpath = '//*[@id="nationGubun"]'
        area = Select(driver.find_element(By.XPATH,xpath))
        area.select_by_visible_text(s[1])

        # 조회버튼 클릭
        xpath = '//*[@id="realContents"]/div/div[2]/div[1]/div[1]/a[2]'
        submit = driver.find_element(By.XPATH, xpath)
        driver.execute_script('arguments[0].click();',submit)
        
        time.sleep(0.5)

        # 값 가져오기
        xpath = '//*[@id="mySheet-table"]/tbody/tr[3]/td/div/div[1]/table/tbody/tr[2]/td[6]/b/span'
        value = driver.find_element(By.XPATH, xpath)
        res.append([date, s[0], s[1], value.text])
```

```python
res
```

```python
df_fin = pd.DataFrame(res)
```

```python
df_fin.rename(columns={0:'년월', 1:'지역', 2:'국가명', 3:'화물(톤)'}, inplace=True)
```

```python
df_fin
```

```python
df_fin['지역'] = df_fin['지역'].replace(['중국'],'아시아')
```

```python
df_fin['지역'] = df_fin['지역'].replace(['일본'],'아시아')
```
